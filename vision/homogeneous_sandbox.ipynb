# Setup

```python
def produce_coordinates(img_data):
    euc_image_inds = np.indices(img_data.shape)
    coord_offset = (np.max(np.max(euc_image_inds, axis=1), axis=1) / 2).astype(int)

    euc_image_coords = euc_image_inds.copy()[::-1, :, :]
    euc_image_coords[1, :, :] = euc_image_coords[1, :, :][::-1]
    euc_image_coords[0] = euc_image_coords[0] - coord_offset[0]
    euc_image_coords[1] = euc_image_coords[1] - coord_offset[1]

    hc_image_coords = np.append(
        euc_image_coords,
        np.ones((1, *img_data.shape), dtype=euc_image_coords.dtype),
        axis=0
    )
    wide_hc_coords = hc_image_coords \
        .reshape((hc_image_coords.shape[0], hc_image_coords.shape[1] * hc_image_coords.shape[2]))

    return euc_image_inds, euc_image_coords, hc_image_coords, wide_hc_coords


def apply_transformation(
    transformation_matrix, euc_image_inds, euc_image_coords,
    hc_image_coords, wide_hc_coords, img_data
    ):
    rotated_hc_coords = np.dot(transformation_matrix, wide_hc_coords) \
        .reshape(hc_image_coords.shape)[:2, :, :] \
        .round() \
        .astype(int)

    coord_shift = rotated_hc_coords - euc_image_coords
    coord_shift[1, :, :] = -1 * coord_shift[1, :, :]
    coord_shift = coord_shift[::-1, :, :]

    rotated_hc_indices = euc_image_inds + coord_shift

    trans_img_data = np.empty(img_data.shape, dtype=img_data.dtype) + 255
    for i in range(img_data.shape[0]):
        for j in range(img_data.shape[1]):
            ind_u = rotated_hc_indices[0, i, j]
            ind_v = rotated_hc_indices[1, i, j]

            in_view = (0 <= ind_u < img_data.shape[0]) and (0 <= ind_v < img_data.shape[1])

            if in_view:
                trans_img_data[ind_u, ind_v] = img_data[i, j]

    trans_img_data = cv2.morphologyEx(
        trans_img_data, cv2.MORPH_OPEN, np.ones((3,3), trans_img_data.dtype)
    )

    return trans_img_data


def run_transform_and_inversion(img_data, transform_matrix):
    euc_image_inds, euc_image_coords, hc_image_coords, wide_hc_coords = produce_coordinates(img_data)
    transformed_image = apply_transformation(
        transform_matrix, euc_image_inds, euc_image_coords,
        hc_image_coords, wide_hc_coords, img_data
    )

    plt.figure(figsize=(7, 7))

    plt.subplot(221)
    plt.imshow(img_data, cmap="Greys_r")
    plt.subplot(222)
    plt.imshow(transformed_image, cmap="Greys_r")

    inv_transform_matrix = np.linalg.inv(transform_matrix)
    euc_image_inds, euc_image_coords, hc_image_coords, wide_hc_coords = produce_coordinates(transformed_image)
    inv_transformed_image = apply_transformation(
        inv_transform_matrix, euc_image_inds, euc_image_coords,
        hc_image_coords, wide_hc_coords, transformed_image
    )

    plt.subplot(223)
    plt.imshow(img_data, cmap="Greys_r")
    plt.subplot(224)
    plt.imshow(inv_transformed_image, cmap="Greys_r")
    
```

# Transformations

## Translation

```python
translation_matrix = np.array([[1, 0, -300], [0, 1, 200], [0, 0, 1]])
run_transform_and_inversion(april_pure_1, translation_matrix)
```

## Rotation

```python
rotation_matrix = np.identity(3)
angle_of_rotation = np.deg2rad(-45)
cos_of_angle = np.cos(angle_of_rotation)
sin_of_angle = np.sin(angle_of_rotation)
rotation_base = np.array([[cos_of_angle, -1 * sin_of_angle], [sin_of_angle, cos_of_angle]])
rotation_matrix[:2, :2] = rotation_base
```

```python
run_transform_and_inversion(april_pure_1, rotation_matrix)
```

## Rigid Body

```python
rigid_body_matrix = np.identity(3)

translation_vector = np.array([-300, 200])

angle_of_rotation = np.deg2rad(-45)
cos_of_angle = np.cos(angle_of_rotation)
sin_of_angle = np.sin(angle_of_rotation)
rotation_base = np.array([[cos_of_angle, -1 * sin_of_angle], [sin_of_angle, cos_of_angle]])

rigid_body_matrix[:2, 2] = translation_vector
rigid_body_matrix[:2, :2] = rotation_base
```

```python
run_transform_and_inversion(april_pure_1, rigid_body_matrix)
```

## Similarity

```python
horizontal_translation = -300
vertical_translation = 200
rotation_angle = -45
scale = 0.5

similarity_matrix = np.identity(3)

translation_vector = np.array([horizontal_translation, vertical_translation])

angle_of_rotation = np.deg2rad(rotation_angle)
cos_of_angle = np.cos(angle_of_rotation)
sin_of_angle = np.sin(angle_of_rotation)
rotation_base = np.array([[cos_of_angle, -1 * sin_of_angle], [sin_of_angle, cos_of_angle]])

rotation_base = scale * rotation_base

similarity_matrix[:2, 2] = translation_vector
similarity_matrix[:2, :2] = rotation_base
```

```python
run_transform_and_inversion(april_pure_1, similarity_matrix)
```

## Affine

```python
horizontal_translation = -300
vertical_translation = 200
origin_rotation_angle = -15
horizontal_scale = 0.5
vertical_scale = 0.5
horizontal_shear_angle = 25
vertical_shear_angle = 25

affine_order = ["T", "R", "SC", "SH"]

origin_rotation_radians = np.deg2rad(origin_rotation_angle)
horizontal_rotation_radians = np.deg2rad(horizontal_shear_angle)
vertical_rotation_radians = np.deg2rad(vertical_shear_angle)

affine_options = {
    "T": np.array(
        [[1, 0, horizontal_translation], [0, 1, vertical_translation], [0, 0, 1]]
    ),
    "R": np.array(
        [
            [np.cos(origin_rotation_radians), -1 * np.sin(origin_rotation_radians), 0],
            [np.sin(origin_rotation_radians), np.cos(origin_rotation_radians), 0],
            [0, 0, 1]
        ]
    ),
    "SC": np.array(
        [[horizontal_scale, 0, 0], [0, vertical_scale, 0], [0, 0, 1]]
    ),
    "SH": np.array(
        [
            [1, np.tan(horizontal_rotation_radians), 0],
            [np.tan(vertical_rotation_radians), 1, 0],
            [0, 0, 1]
        ]
    )
}

affine_chain = [affine_options[choice] for choice in affine_order]
affine_matrix = affine_chain[0]
for affine_option in affine_chain[1:]:
    affine_matrix = np.dot(affine_matrix, affine_option)

```

```python
run_transform_and_inversion(april_pure_1, affine_matrix)
```

## Projective

```python
horizontal_translation = 100
vertical_translation = 0
origin_rotation_angle = 0
horizontal_scale = 1
vertical_scale = 1
horizontal_shear_angle = 0
vertical_shear_angle = 0
projection_x = 10
projection_y = 2

affine_order = ["T", "R", "SC", "SH"]

origin_rotation_radians = np.deg2rad(origin_rotation_angle)
horizontal_rotation_radians = np.deg2rad(horizontal_shear_angle)
vertical_rotation_radians = np.deg2rad(vertical_shear_angle)

affine_options = {
    "T": np.array(
        [[1, 0, horizontal_translation], [0, 1, vertical_translation], [0, 0, 1]]
    ),
    "R": np.array(
        [
            [np.cos(origin_rotation_radians), -1 * np.sin(origin_rotation_radians), 0],
            [np.sin(origin_rotation_radians), np.cos(origin_rotation_radians), 0],
            [0, 0, 1]
        ]
    ),
    "SC": np.array(
        [[horizontal_scale, 0, 0], [0, vertical_scale, 0], [0, 0, 1]]
    ),
    "SH": np.array(
        [
            [1, np.tan(horizontal_rotation_radians), 0],
            [np.tan(vertical_rotation_radians), 1, 0],
            [0, 0, 1]
        ]
    )
}

affine_chain = [affine_options[choice] for choice in affine_order]
projection_matrix = affine_chain[0]
for affine_option in affine_chain[1:]:
    projection_matrix = np.dot(projection_matrix, affine_option)

projection_matrix[2, :2] = np.array([projection_x, projection_y])
```

```python
run_transform_and_inversion(april_pure_1, projection_matrix)
```
